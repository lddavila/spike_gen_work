```python
def plot_n_channels(rec_gen_object,channels,single_plot_or_not,figure_file_name_save,y_limits):
    time_stamps = rec_gen_object.timestamps[:]
    if single_plot_or_not:
        fig, ax = plt.subplots(len(channels),figsize=(40,40))
        for i in range(len(channels)):
            voltage_readings = rec_gen_object.recordings[:,chLannels[i]]
            #print(rec_gen_object.timestamps.size)
            #print(rec_gen_object.recordings[:,channels[i]].size)
            ax[i].plot(time_stamps,voltage_readings)
            ax[i].set_xlabel("Time in seconds")
            ax[i].set_ylabel("Voltage")
            ax[i].set_ylim(y_limits)
            ax[i].set_title("Channel "+ str(channels[i]))
            
        plt.show()
        
    else:
        fig, ax = plt.subplots()
        for i in range(len(channels)):
            plt.plot(time_stamps,rec_gen_object.recordings[:,channels[i]],label="Channel"+str(i))
        plt.xlabel("Time in seconds")
        plt.ylabel("Voltage")
        plt.title("Channels")
        plt.legend()
        plt.show()
    fig.savefig(figure_file_name_save)
```

```python
def save_recordings_by_channel(recording_matrix,dir_to_save_things_to):
    os.mkdir(dir_to_save_things_to)
    current_directory = os.getcwd();
    os.chdir(dir_to_save_things_to)
    for i in range(384):
        channel_recording = {"c"+str(i+1):recording_matrix[:,i]};
        savemat("c"+str(i+1)+".mat",channel_recording);
    os.chdir(current_directory);
```
